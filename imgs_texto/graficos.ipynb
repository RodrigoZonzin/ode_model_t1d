```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.colors as mcolors
from matplotlib import colormaps
import random
```

```python
# Parâmetros
k = 1.0
m = 2.0
```

```python
# EDO: dy/dt = ky + m
def dydt(t, y):
    return k +k*y

# Condição inicial
y0 = [0]
```

```python
t_span = (0, 5)
t_eval = np.linspace(*t_span, 100)
fig = plt.figure(figsize=(10, 6), dpi = 300)

cmap_vir = colormaps['inferno']


mycolors = cmap_vir(np.linspace(0, 1, 20))

for i, (dx, dy) in enumerate([(np.random.rand(), np.random.rand()) for _ in range(20)]):
    k = 0.0001
    m = 0.0008
    k = k+dx/2
    m = m+dy/2
    sol = solve_ivp(dydt, t_span, y0, t_eval=t_eval)
    plt.plot(sol.t, sol.y[0], label=f'k={k:.4f}, m = {m:.4f}', color = mycolors[i], alpha = 0.96)

plt.xlabel('t')
plt.ylabel(r'$y(t)$')
plt.legend()
plt.tight_layout()
plt.savefig('result.png')
```

```python
f = lambda a, x, t: a*x + t
g = lambda a, x: np.exp(a*x)

cmap_vir = colormaps['viridis']
mycolors = cmap_vir(np.linspace(0, 1, 30))
    

#fig, axs = plt.subplots(*(1, 2), figsize=(16, 8), dpi = 300)
fig = plt.figure(figsize=(8, 8), dpi= 300)

for i, a in enumerate(range(0, 25, 1)):    
    x = np.linspace(0, 50)
    fx = f(a, x, 0)
    plt.plot(fx, label = f'f(x)={a}x+5', color = mycolors[i])

ax = plt.gca()
ax.text(0.02, 0.98, r'$\frac{dy}{dt} = \alpha $', transform=ax.transAxes,
            verticalalignment='top', horizontalalignment='left',
            fontsize = 20,
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.1'))

plt.tight_layout()
plt.savefig("EDOConstante.png")


fig = plt.figure(figsize=(8, 8), dpi= 300)
values = np.sort(np.random.uniform(size = 25))/15

for i, a in enumerate(values):
    x = np.linspace(0, 50)
    gx = g(a, x)
    plt.plot(gx, label = f'g(x) = e^{a}x', color = mycolors[i])

ax = plt.gca()
ax.text(0.02, 0.98, r'$\frac{dy}{dt} = \beta $y', transform=ax.transAxes,
            verticalalignment='top', horizontalalignment='left',
            fontsize = 20,
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.1'))

plt.tight_layout()
plt.savefig("EDOLinear.png")



fig = plt.figure(figsize=(8, 8), dpi= 300)
values = np.sort(np.random.uniform(size = 25))/15

for i, a in enumerate(values):
    x = np.linspace(0, 50)
    lgx = lg(10, 1, 0.5, x)
    plt.plot(gx, label = f'g(x) = e^{a}x', color = mycolors[i])

ax = plt.gca()
ax.text(0.02, 0.98, r'$\frac{dy}{dt} = rP (1 - \frac{P}{K}) $y', transform=ax.transAxes,
            verticalalignment='top', horizontalalignment='left',
            fontsize = 20,
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.1'))

plt.tight_layout()
plt.savefig("EDOLogistica.png")
```
