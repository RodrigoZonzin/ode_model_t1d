```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.colors as mcolors
from matplotlib import colormaps
import random
```

```python
# Parâmetros
k = 1.0
m = 2.0
```

```python
# EDO: dy/dt = ky + m
def dydt(t, y):
    return k +k*y

# Condição inicial
y0 = [0]
```

```python
t_span = (0, 5)
t_eval = np.linspace(*t_span, 100)
fig = plt.figure(figsize=(10, 6), dpi = 300)

cmap_vir = colormaps['inferno']


mycolors = cmap_vir(np.linspace(0, 1, 20))

for i, (dx, dy) in enumerate([(np.random.rand(), np.random.rand()) for _ in range(20)]):
    k = 0.0001
    m = 0.0008
    k = k+dx/2
    m = m+dy/2
    sol = solve_ivp(dydt, t_span, y0, t_eval=t_eval)
    plt.plot(sol.t, sol.y[0], label=f'k={k:.4f}, m = {m:.4f}', color = mycolors[i], alpha = 0.96)

plt.xlabel('t')
plt.ylabel(r'$y(t)$')
plt.legend()
plt.tight_layout()
plt.savefig('result.png')
```

```python
f = lambda a, x, t: a*x + t
g = lambda m, k, t: (m/k)*(np.exp(k*t)-1)
```

```python
cmap_vir = colormaps['inferno']
mycolors = cmap_vir(np.linspace(0, 3))
x = np.linspace(0, 50)

fig, axs = plt.subplots(*(1, 2), figsize=(16, 8), dpi = 300)

for i, a in enumerate(range(0, 25, 1)):    
    fx = f(a, x, 0)
    axs[0].plot(fx, label = f'f(x)={a}x+5', color = mycolors[i])

for i, (dx, dy) in enumerate([(np.random.rand(), np.random.rand()) for _ in range(20)]):
    k = 0.0001
    m = 0.0008
    k = k+dx/2
    m = m+dy/2

    gx = g(m, k, x)
    axs[1].plot(gx, color = mycolors[i])


axs[0].text(0.02, 0.98, r'+m', transform=axs[0].transAxes,
            verticalalignment='top', horizontalalignment='left',
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

axs[1].text(0.02, 0.98, r'+ky', transform=axs[1].transAxes,
            verticalalignment='top', horizontalalignment='left',
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

plt.xlim(0, 20)
plt.ylim(0, 200)
plt.tight_layout()
#plt.show()
plt.savefig("exemploEDOs.png")
#axs[0].legend()
```
